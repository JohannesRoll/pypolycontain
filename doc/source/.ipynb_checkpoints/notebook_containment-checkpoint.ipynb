### Example 2 on Necessary and Sufficiency for Containment 

This time we look for zonotope inside V-polytope

```python
import numpy as np
np.random.seed(0)
```

```python
from pypolycontain.objects import zonotope,V_polytope,unitbox
from pypolycontain.containment import subset
from pypolycontain.conversions import to_V,to_AH_polytope
from pypolycontain.visualize import visualize
```

Import mathematical program with Gurobi solver from pydrake. 
_**Warning:** Here I assume you have Gurobi solver installed with Drake's python bindings. Otherwise, replace Gurobi with SCS solver below.

```python
import pydrake.solvers.mathematicalprogram as MP
import pydrake.solvers.gurobi as Gurobi_drake
# use Gurobi solver
global gurobi_solver, license
gurobi_solver=Gurobi_drake.GurobiSolver()
license = gurobi_solver.AcquireLicense()
```

```python
V_circimbody=V_polytope([10*np.random.random((2,1))-5 for i in range(18)])
circumbody=to_AH_polytope(V_circimbody)
circumbody.color='blue'
Z_inbody=zonotope(x=np.zeros((2,1)),G=np.random.random((2,5))-0.5,color='red')
visualize([circumbody,Z_inbody])
```

### Optimization
Now we solve the following optimization problem

```python
s_n,N={},20
for i in range(N):
    prog=MP.MathematicalProgram()
    s=prog.NewContinuousVariables(1,"s")
    Z_s=zonotope(Z_inbody.x,Z_inbody.G*s)
    prog.AddLinearCost(-np.array([1]),np.array([0]),s)
    subset(prog,Z_s,circumbody,N=i-1)
    result=gurobi_solver.Solve(prog,None,None)
    if result.is_success():
        print("subset test successfull")
        s_n[i]=result.GetSolution(s)
        print(i-1,"s=",s_n[i])
    else:
        print("not a subset") 
```

```python
visualize([circumbody,zonotope(Z_inbody.x,Z_inbody.G*(s_n[1]))])
```

```python
import matplotlib.pyplot as plt
plt.plot(range(N),[s_n[i] for i in range(N)])
plt.plot(range(N),[s_n[i] for i in range(N)],'o')
plt.title("Less is more conservative")
```
