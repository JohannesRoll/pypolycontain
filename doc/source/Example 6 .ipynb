# Example 6: Hausdorff Distance

```python
import numpy as np
import pypolycontain as pp
import time
import pydrake.solvers.mathematicalprogram as MP
# use Gurobi solver
import pydrake.solvers.gurobi as Gurobi_drake
gurobi_solver=Gurobi_drake.GurobiSolver()
np.random.seed(13)
```

```python
X=pp.zonotope(x=np.zeros((2,1)),G=np.random.randint(-3,4,size=(2,3)),color='blue')
Y=pp.zonotope(x=np.ones((2,1)),G=np.random.randint(-3,4,size=(2,3)),color='red')
pp.visualize([X,Y])
print(X.G)
print(Y.G)
```

```python
ball=pp.unitball(2,1)
D,D_12,D_21=pp.Hausdorff_distance(X,Y,ball=1)
X_plus=pp.minkowski_sum(X,D_12*ball)
Y_plus=pp.minkowski_sum(Y,D_21*ball)
X_plus.color='cyan'
Y_plus.color='pink'
```

```python
pp.visualize([X_plus,X,Y],title=r'$\vec{d}_H(\mathbb{X},\mathbb{Y})=%0.02f$'%D_12,equal_axis=True,TitleSize=30)
pp.visualize([Y_plus,X,Y],title=r'$\vec{d}_H(\mathbb{Y},\mathbb{X})=%0.02f$'%D_21,equal_axis=True,TitleSize=30)
```
