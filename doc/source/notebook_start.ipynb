# Getting Started

The simplest task in `pypolycontain` is defining polytopic objects, performing operations, and visualize them. First, we import the package alongside `numpy`.

````python
import numpy as np
import pypolycontain as pp
````

## Objects
### First H-polytope
We define an H-polytope $\mathbb{P}=\{x \in \mathbb{R}^2 | Hx \le h\}$. We give the following numericals:

````python
H=np.array([[1,1],[-1,1],[0,-1]])
h=np.array([[1,1,0]]).reshape(3,1)
my_triangle=pp.objects.H_polytope(H,h)
````

In order to visualizate the polytope, we call the following function. Note the brackets around `pypolycontain` as the `visualize` function takes in a list of polytopes as its primary argument. 

````python
pp.visualize.visualize([my_triangle])
````

### AH-polytope
We define an AH-polytope as $t+T\mathbb{P}$ with the following numbers that represents a rotation of $30^\circ$ and translation in $x$ direction.

````python
t=np.array([5,0]).reshape(2,1) # offset
theta=np.pi/6 # 30 degrees
T=np.array([[np.cos(theta),np.sin(theta)],[-np.sin(theta),np.cos(theta)]]) # Linear transformation
my_AH_polytope=pp.objects.AH_polytope(t,T,my_triangle)
pp.visualize.visualize([my_AH_polytope])
````

### Zonotope
We define a zonotope as $\mathbb{Z}=x+G[-1,1]^{n_p}$, where $n_p$ is the number of rows in $p$. 

````python
x=np.array([3,0]).reshape(2,1) # offset
G=np.array([[1,0,1],[0,0,-1]]).reshape(2,3)
my_zonotope=pp.objects.zonotope(x,G)
pp.visualize.visualize([my_zonotope])
````

## Visualization
The `visualize` function allows for visualizing multiple polytopic objects. The default color for ```H-polytope``` is red, ```zonotope``` is green, and ```AH-polytope``` is blue.

````python
pp.visualize.visualize([my_triangle,my_AH_polytope,my_zonotope],title="All of My Polytopes")
````

## Operations

### Minkowski sum

````python
sum_polytope=pp.operations.minkowski_sum(my_triangle,my_zonotope)
sum_polytope.color=(0.6,0.6,0)
pp.visualize.visualize([sum_polytope])
````
